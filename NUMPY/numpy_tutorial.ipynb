```python
import numpy as np
import time
```

```python
arr = np.array([1,2,3,4,5])
print(arr, type(arr))
```

```python
size = 1_000_000

# python list
py_list = list(range(size))
start = time.time()
sq_list = [x**2 for x in py_list]
end = time.time()
print(f"python list time : {end-start} sec")

# numpy arr
np_arr = np.array(py_list)
start = time.time()
sq_array = np_arr ** 2
end = time.time()
print(f"numpy array time : {end-start} sec")
```

```python
# memory
import sys

print(f"python list size = {sys.getsizeof(py_list)* len(py_list)} byte")

print(f"numpy array size = {np_arr.nbytes} byte")
```
