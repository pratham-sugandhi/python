```python
# cleaning data --

df = pd.read_csv("raw_data.csv")
df

# missing val-

df.isnull()    #returns true where data is missing
df.isna()

df.isnull().sum()

df.dropna()  #drop null cell row 
df.dropna(axis = 1)

df.fillna(0)

cleaned_df =  df.copy()
age_mean = cleaned_df["age"].mean()

cleaned_df["age"] = cleaned_df["age"].fillna(age_mean)

df.ffill()
df.bfill()
```
