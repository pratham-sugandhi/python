```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
print(type(tips))
tips.head()
```

```python

# relation plot
sns.set_theme(style = "darkgrid")

sns.relplot(
    data = tips,
    x = "total_bill",
    y = "tip",
    hue = "smoker",
    size = "size",
    style = "smoker"
)
```

```python

x_val = [1,2,3,4,5,6]
y_sq = [i**2 for i in x_val]

sns.set_theme(style="ticks")

sns.relplot(
    x = x_val,
    y = y_sq,
    kind = "line"
)
```

```python

tips = sns.load_dataset("tips") 

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python

flight = sns.load_dataset("flights")

sns.lineplot(
    data = flight,
    x = "year",
    y = "passengers"
)
```

```python

tips = sns.load_dataset("tips") 

sns.barplot(
    data = tips,
    x = "day",
    y = "tip",
    hue = "sex"
)
```

```python

sns.boxplot(
    data = tips,
    x = "day",
    y = "tip",
    hue = "sex"
)
```

```python

penguins = sns.load_dataset("penguins")

sns.histplot(
    data = penguins,
    x = "body_mass_g",
    bins = 30
)
```

```python
# heatmaps- matrix form for rect dataset

flights_pivot = flight.pivot(index = "month", columns="year", values="passengers")

sns.heatmap(
    flights_pivot,
    cmap = "coolwarm",
    annot = True,
    fmt = "d"
)

plt.title("passengers heatmap")
plt.tight_layout()
```

```python
# pract -

fig , ax = plt.subplots()

sns.lineplot(
    data = tips,
    x = "day",
    y = "total_bill",
    hue = "sex",
    marker = "o",
    ax = ax,
    errorbar=None
)

ax.set_title("avg bill value")
ax.set_xlabel("Days")
ax.set_ylabel("Total bill amount")

fig.tight_layout()
```
