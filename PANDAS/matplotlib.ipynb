```python
# real data--

years = [2011, 2014, 2017, 2020, 2023]
oscar_award = [1003, 1800, 1200, 2500, 2000]

plt.plot(years, oscar_award)
plt.title("Num of oscar awards winner!")
plt.xlabel("Years")
plt.ylabel("No of winners")
```

```python

non_oscar = [3335, 4563, 1910, 2199, 3999]

plt.plot(years, oscar_award, "^--r", label = "Oscar Movies")
plt.plot(years, non_oscar, "o-b", label = "Non Oscar Movies")

plt.title("Num of oscar vs non oscar awards")
plt.xlabel("years")
plt.ylabel("Num of winners")
plt.legend(loc = "upper right")
```

```python


years = [2010, 2014, 2018, 2022, 2026]
oscar_award = [1003, 1800, 1200, 2500, 2000]

plt.bar(years, oscar_award, color = "r")
plt.title("Num of oscar awards in years")
plt.xlabel("Years")
plt.ylabel("Num of oscar winner")
plt.legend()

for i in range (len(years)):
    plt.text(years[i], oscar_award[i]+30, str(oscar_award[i]), ha="center")

plt.ylim(0 , max(oscar_award)+150)
```

```python
import numpy as np

oscar_movies = ["Dark", "The Artist", "King SRK", "Avatar", "Dhoom"]
non_oscar = ["light", "Bhoot", "Bangla", "toxic", "Butter"]

years = [2010, 2014, 2018, 2022, 2026]

oscar_award = [1003, 1800, 1300, 2500, 2000]
non_oscar_award = [923, 1200, 2100, 1400, 1700]

x = np.arange(len(years))
width = 0.4
plt.bar(x-width/2, oscar_award, width, label="oscar award")
plt.bar(x+width/2, non_oscar_award, width, label="non oscar award")

plt.title("Oscar vs Non Oscar")
plt.xlabel("Years")
plt.ylabel("Awards")

plt.legend()
plt.xticks(x, years)
```
